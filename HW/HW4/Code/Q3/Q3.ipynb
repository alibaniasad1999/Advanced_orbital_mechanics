```python
# Qesution 3 #
day2rep = 6.875
rev2rep = 107
revPday = rev2rep/day2rep
e = 0
mu = 398600
omega_e = 2*np.pi/86400
n = revPday*omega_e
R_e = 6378
j2 = 0.00108263
Omega_dot_sun_synchronous = 2*np.pi/(365.25*86400)
```

```python
# part a #
a_n = (mu*(1/n)**2)**(1/3)
# algorithm 71 #
delta_landa_rev = 2*np.pi*day2rep/rev2rep
a = 0
while abs(a-a_n) > 0.001:
    a = a_n
    print(a)
    p = a*(1-e**2)
    delta_landa_period = 2*np.pi*Omega_dot_sun_synchronous/n
    delta_lon = delta_landa_rev + delta_landa_period
    n = 2*np.pi*omega_e/delta_lon
    a_n = (mu*(1/n)**2)**(1/3)
print('a =',a_n,'km')
# calculate i possible #
i = np.arccos(-a**(7/2)*Omega_dot_sun_synchronous/(3*j2*(R_e)**2*np.sqrt(mu)))
print('i =',i*180/np.pi,'deg')
print(-a**(7/2)*Omega_dot_sun_synchronous/(3*j2*(R_e)**2*np.sqrt(mu)))
```

```python
def init_a(i, Omega_dot_sun_synchronous, n, j2, R_e):
    a = np.sqrt(-3/2*n*j2*R_e**2*np.cos(i)/Omega_dot_sun_synchronous)

    return a
```

```python
i_array = np.linspace(90*np.pi/180, 180*np.pi/180, 100)
for i  in i_array:
    a_n = init_a(i, Omega_dot_sun_synchronous, n, j2, R_e)
    a = 0
    print('a init =',a_n,'km')
    while abs(a-a_n) > 0.001:
        a = a_n
        p = a*(1-e**2)
        delta_landa_period = 2*np.pi*Omega_dot_sun_synchronous/n
        delta_lon = delta_landa_rev + delta_landa_period
        n = 2*np.pi*omega_e/delta_lon
        a_n = (mu*(1/n)**2)**(1/3)
    # calculate i possible #
    print('a ans: ', a_n)
    i_ans = np.arccos(-a_n**(7/2)*Omega_dot_sun_synchronous/(3*j2*(R_e)**2*np.sqrt(mu)))
    print('i init =',i*180/np.pi,'deg')
    print('i ans =',i_ans*180/np.pi,'deg')
    # print(-a_n**(7/2)*Omega_dot_sun_synchronous/(3*j2*(R_e)**2*np.sqrt(mu)))
        
    
```
