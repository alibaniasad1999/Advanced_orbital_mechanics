```python
# part d #
e = np.linspace(0,0.9, 1000)
i = np.linspace(90,180, 1000)
a = np.zeros((len(e),len(i)))
for j in range(len(e)):
    for k in range(len(i)):
        a[j,k] = (-(3/2*R**2*J2*np.sqrt(mu)*np.cos(i[k]*np.pi/180))/(dOmega*(1-e[j]**2)**2))**(2/7)
```

```python
# relation bewteen a and r pregee #
r_p = a*(1-e**2)/(1+e)
```

```python
import math

def calculate_orbit_parameters(mu, r, v):
    # Calculate specific angular momentum
    h = r * v

    # Calculate eccentricity vector
    e_vector = [(v[1]*h[2] - v[2]*h[1]) / mu,
                (v[2]*h[0] - v[0]*h[2]) / mu,
                (v[0]*h[1] - v[1]*h[0]) / mu]
    
    # Calculate eccentricity magnitude
    e = math.sqrt(e_vector[0]**2 + e_vector[1]**2 + e_vector[2]**2)
    
    # Calculate semi-major axis
    a = r / (2 - r * v.dot(v) / mu)
    
    # Calculate inclination
    i = math.degrees(math.acos(h[2] / math.sqrt(h[0]**2 + h[1]**2 + h[2]**2)))
    
    return i, e, a

# Example inputs
mu = 3.24859e5  # Venus gravitational parameter in km^3/s^2
r = [10000, 20000, 30000]  # Position vector in km
v = [-1, 2, -3]  # Velocity vector in km/s

# Calculate orbit parameters
inclination, eccentricity, semi_major_axis = calculate_orbit_parameters(mu, r, v)

print("Inclination:", inclination, "degrees")
print("Eccentricity:", eccentricity)
print("Semi-major Axis:", semi_major_axis, "km")
```
