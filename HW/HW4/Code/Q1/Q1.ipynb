```python
## Question One ##
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Sun-synchronous orbits ##
# Constants
mu = 398600.4415 # km^3/s^2
J2 = 0.00108248
R = 6378.137 # km
a = R+600 # km
dOmega = 2*np.pi/(365.25*24*3600) # rad/s
```

```python
# part a #
e = 0

i = np.arccos(-2/3*a**(7/2)*dOmega*(1-e**2)**2/(R**2*J2*np.sqrt(mu))) # rad
print('e = 0, i =',np.rad2deg(i),'deg')
```

```python
# part b #
e = 0.1

i = np.arccos(-2/3*a**(7/2)*dOmega*(1-e**2)**2/(R**2*J2*np.sqrt(mu))) # rad
print('e = 0, i =',np.rad2deg(i),'deg')
```

```python
# part c #
e = np.linspace(0,0.9, 1000)
a = np.linspace(400,1000, 1000)
data = np.zeros((len(e),len(a)))

for i in range(len(e)):
    for j in range(len(a)):
        data[i,j] = np.arccos(-2/3*a[j]**(7/2)*dOmega*(1-e[i]**2)**2/(R**2*J2*np.sqrt(mu)))
```

```python
# plot 3D #
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(a, e)
ax.plot_surface(X, Y, data*180/np.pi, cmap='viridis')
ax.set_xlabel('a (km)')
ax.set_ylabel('e')
ax.set_zlabel('i (deg)')
# save eps #
fig.savefig('myimage.svg', format='svg', dpi=1200)
plt.show()
```
